```python
from transformers import pipeline

sentiment_classifier = pipeline("sentiment-analysis")
```

```python
words = "I am so happy to start learning AI today!"
results = sentiment_classifier(words)

print(f"Words: {words}")
print(f"Results: {results}")
```

```python
reviews = [
    "The product is amazing, I love it!",
    "Terrible experience, the item arrived broken.",
    "It's okay, not bad but not great either.",
    "Fast delivery and high quality."
]

for review in reviews:
    res = sentiment_classifier(review)[0]
    print(f"Review: {review}")
    print(f"Sentiment: {res['label']} (Score: {res['score']:.2f})")
    print("-" * 20)
```
